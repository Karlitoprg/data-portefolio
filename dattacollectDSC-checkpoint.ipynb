```python
import pandas as pd
```

```python
da1 = pd.read_excel("dattacollect_amazon.excel.xlsx")
da2 = pd.read_excel("dattacollect_back.excel.xlsx")
```

```python
# Remplacez les espaces par des points dans la colonne "modele" du DataFrame da1
da1['modele'] = da1['modele'].str.replace(' ', '.')

# Remplacez les espaces par des points dans la colonne "modele" du DataFrame da2
da2['modele'] = da2['modele'].str.replace(' ', '.')
```

```python
da1['modele']
da2['modele']
```

```python
# Supprimez les lignes vides du DataFrame da1
da1 = da1.dropna()

# Supprimez les lignes vides du DataFrame da2
da2 = da2.dropna()
```

```python
# Supprimez les lignes en double du DataFrame da1
da1 = da1.drop_duplicates()

# Supprimez les lignes en double du DataFrame da2
da2 = da2.drop_duplicates()
```

```python
import pandas as pd

# Supposons que "da1" est notre DataFrame
da1['prix'] = pd.to_numeric(da1['prix'], errors='coerce')

# Vérifions si toutes les valeurs ont été converties avec succès
if da1['prix'].notnull().all():
    print("La colonne 'prix' est constituée de données numériques.")
else:
    print("La colonne 'prix' contient des données non numériques.")
```

```python
# Supprimons les lignes avec des valeurs non numériques dans la colonne "prix"
da1 = da1.dropna(subset=['prix'])
```

```python
import pandas as pd

# Supposons que "da1" est votre DataFrame
da2['prix'] = pd.to_numeric(da2['prix'], errors='coerce')

# Vérifiez si toutes les valeurs ont été converties avec succès
if da2['prix'].notnull().all():
    print("La colonne 'prix' est constituée de données numériques.")
else:
    print("La colonne 'prix' contient des données non numériques.")
```

```python
# Statistiques descriptives pour toutes les colonnes de da1
print(da1.describe())

# Statistiques descriptives pour toutes les colonnes de da2
print(da2.describe())
```

```python
import matplotlib.pyplot as plt

# Créez un histogramme pour les avis dans le premier DataFrame
plt.hist(da1['avis'], bins=10, edgecolor='k')
plt.xlabel('Avis')
plt.ylabel('Nombre d\'occurrences')
plt.title('Distribution des Avis dans da1')
plt.show()

# Créez un histogramme pour les avis dans le deuxième DataFrame
plt.hist(da2['avis'], bins=10, edgecolor='k')
plt.xlabel('Avis')
plt.ylabel('Nombre d\'occurrences')
plt.title('Distribution des Avis dans da2')
plt.show()

# Créez un histogramme pour les prix dans le premier DataFrame
plt.hist(da1['prix'], bins=20, edgecolor='k')
plt.xlabel('Prix')
plt.ylabel('Nombre d\'occurrences')
plt.title('Distribution des Prix dans da1')
plt.show()

# Créez un histogramme pour les prix dans le deuxième DataFrame
plt.hist(da2['prix'], bins=20, edgecolor='k')
plt.xlabel('Prix')
plt.ylabel('Nombre d\'occurrences')
plt.title('Distribution des Prix dans da2')
plt.show()
```

```python
# Créez un nuage de points pour examiner la relation entre les avis et les prix dans da1
plt.scatter(da1['avis'], da1['prix'])
plt.xlabel('Avis')
plt.ylabel('Prix')
plt.title('Nuage de Points : Avis vs. Prix dans da1')
plt.show()

# Créez un nuage de points pour examiner la relation entre les avis et les prix dans da2
plt.scatter(da2['avis'], da2['prix'])
plt.xlabel('Avis')
plt.ylabel('Prix')
plt.title('Nuage de Points : Avis vs. Prix dans da2')
plt.show()
```

```python
# Pour le DataFrame da1
modeles_populaires_da1 = da1['modele'].value_counts()  # Comptez le nombre d'occurrences de chaque modèle
moyenne_prix_da1 = da1.groupby('modele')['prix'].mean()  # Calculez la moyenne des prix pour chaque modèle

# Pour le DataFrame da2
modeles_populaires_da2 = da2['modele'].value_counts()  # Comptez le nombre d'occurrences de chaque modèle
moyenne_prix_da2 = da2.groupby('modele')['prix'].mean()  # Calculez la moyenne des prix pour chaque modèle
```

```python
Q1 = da1['prix'].quantile(0.25)
Q3 = da1['prix'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = da1[(da1['prix'] < lower_bound) | (da1['prix'] > upper_bound)]
```

```python
# Pour le DataFrame da1
modeles_populaires_da1 = da1['modele'].value_counts().head(10).index

# Pour le DataFrame da2
modeles_populaires_da2 = da2['modele'].value_counts().head(10).index
```

```python
# Pour le DataFrame da1
da1_modeles_populaires = da1[da1['modele'].isin(modeles_populaires_da1)]

# Pour le DataFrame da2
da2_modeles_populaires = da2[da2['modele'].isin(modeles_populaires_da2)]
```

```python
# Pour le DataFrame da1
moyenne_prix_par_modele_etat_da1 = da1_modeles_populaires.groupby(['modele', 'etat'])['prix'].mean()

# Pour le DataFrame da2
moyenne_prix_par_modele_etat_da2 = da2_modeles_populaires.groupby(['modele', 'etat'])['prix'].mean()
```
